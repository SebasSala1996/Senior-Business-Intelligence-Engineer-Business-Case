```python
import csv
import random
import datetime

# Define the start date and the duration (three months approximated as 90 days)
start_date = datetime.datetime(2025, 1, 1)
duration_days = 90
total_seconds = duration_days * 24 * 60 * 60  # total seconds in 90 days

def partition_integer(total):
    """Randomly partition an integer 'total' into three non-negative integers that sum to 'total'."""
    # Generate two random breakpoints between 0 and total
    r1 = random.randint(0, total)
    r2 = random.randint(0, total)
    # Ensure r1 is the smaller value
    low, high = min(r1, r2), max(r1, r2)
    part1 = low
    part2 = high - low
    part3 = total - high
    return part1, part2, part3

# Open the CSV file for writing
with open("connections.csv", "w", newline="") as csvfile:
    writer = csv.writer(csvfile)

    # Write the header row with the added ID_Date column
    writer.writerow([
        "Connection_ID",
        "Timestamp",
        "ID_Date",
        "User_ID",
        "Session_Duration",
        "Feature_A_Count",
        "Feature_B_Count",
        "Feature_C_Count",
        "Is_Conversion"
    ])

    # Generate 10000 connection rows
    for connection_id in range(1, 10001):
        # Random timestamp within the three-month period
        random_seconds = random.randint(0, total_seconds)
        timestamp = start_date + datetime.timedelta(seconds=random_seconds)
        formatted_timestamp = timestamp.strftime("%Y-%m-%d %H:%M:%S")
        # Generate ID_Date in YYYYMMDD format
        id_date = timestamp.strftime("%Y%m%d")

        # Random user id between 1 and 1000 (assume a pool of 1000 users)
        user_id = random.randint(1, 1000)

        # Session duration as an integer between 5 and 30 minutes
        session_duration = random.randint(5, 30)

        # Partition the session_duration into three parts for feature usage
        feature_a_count, feature_b_count, feature_c_count = partition_integer(session_duration)

        # Conversion flag (10% chance for a conversion)
        is_conversion = 1 if random.random() < 0.10 else 0

        # Write the row to CSV
        writer.writerow([
            connection_id,
            formatted_timestamp,
            id_date,
            user_id,
            session_duration,
            feature_a_count,
            feature_b_count,
            feature_c_count,
            is_conversion
        ])

print("CSV file 'connections.csv' generated successfully.")
```

```python
import csv
import random

# Number of users (set to 1000)
num_users = 1000

# Define sample names for each region
us_first_names = ["James", "John", "Robert", "Michael", "William", "David", "Richard", "Joseph", "Thomas", "Charles"]
us_last_names = ["Smith", "Johnson", "Williams", "Jones", "Brown", "Davis", "Miller", "Wilson", "Moore", "Taylor"]

uk_first_names = ["Oliver", "George", "Harry", "Jack", "Jacob", "Noah", "Charlie", "Muhammad", "Thomas", "Oscar"]
uk_last_names = ["Smith", "Jones", "Williams", "Taylor", "Brown", "Davies", "Evans", "Wilson", "Thomas", "Roberts"]

france_first_names = ["Jean", "Pierre", "Michel", "Alain", "Francois", "Luc", "Andre", "Jacques", "Philippe", "Henri"]
france_last_names = ["Martin", "Bernard", "Dubois", "Thomas", "Robert", "Richard", "Petit", "Durand", "Leroy", "Moreau"]

germany_first_names = ["Hans", "Peter", "Klaus", "Heinz", "Uwe", "Jurgen", "Wolfgang", "Fritz", "Gunter", "Werner"]
germany_last_names = ["Muller", "Schmidt", "Schneider", "Fischer", "Weber", "Meyer", "Wagner", "Becker", "Schulz", "Hoffmann"]

# Dictionary mapping each country to its name lists
name_data = {
    "US": {"first": us_first_names, "last": us_last_names},
    "UK": {"first": uk_first_names, "last": uk_last_names},
    "France": {"first": france_first_names, "last": france_last_names},
    "Germany": {"first": germany_first_names, "last": germany_last_names}
}

countries = list(name_data.keys())

# Open the CSV file for writing
with open("users.csv", "w", newline="", encoding="utf-8") as csvfile:
    writer = csv.writer(csvfile)

    # Write the header row with the new Segment column
    writer.writerow(["User_ID", "First_Name", "Last_Name", "Name", "Paid_User", "Churn", "Segment"])

    # Generate user rows
    for user_id in range(1, num_users + 1):
        # Randomly choose a country
        country = random.choice(countries)
        first_name = random.choice(name_data[country]["first"])
        last_name = random.choice(name_data[country]["last"])
        name = f"{first_name} {last_name}"

        # Assume a 40% chance that a user is paid
        paid_user = 1 if random.random() < 0.40 else 0
        # Assume a 15% chance that a user has churned
        churn = 1 if random.random() < 0.15 else 0

        # Segment users
        if churn == 1:
            segment = "Lapsed User"
        elif paid_user == 1:
            segment = "Power User"
        else:
            segment = "New User"

        writer.writerow([user_id, first_name, last_name, name, paid_user, churn, segment])

print("CSV file 'users.csv' generated successfully.")
```

```python
import csv
import random
import datetime

# Parameters
num_users = 1000
features = ["Feature A", "Feature B", "Feature C"]

# Define the start date and the duration (three months approximated as 90 days)
start_date = datetime.datetime(2025, 1, 1)
duration_days = 90
total_seconds = duration_days * 24 * 60 * 60  # Total seconds in 90 days

# Open the CSV file for writing
with open("user_feature_interactions.csv", "w", newline="", encoding="utf-8") as csvfile:
    writer = csv.writer(csvfile)

    # Write the header row including the new ID_Date column
    writer.writerow(["User_ID", "Feature", "Interaction_Count", "Total_Interaction_Time_Minutes", "Timestamp", "ID_Date"])

    # Generate aggregated interaction metrics for each user and feature
    for user_id in range(1, num_users + 1):
        for feature in features:
            # Generate a random number of interactions (for example, between 0 and 100)
            interaction_count = random.randint(0, 100)
            # For each interaction, assume an average time between 1 and 5 minutes
            if interaction_count > 0:
                avg_time = random.uniform(1, 5)
                total_time = round(interaction_count * avg_time, 1)
            else:
                total_time = 0.0

            # Generate a random timestamp within the three-month period
            random_seconds = random.randint(0, total_seconds)
            timestamp = start_date + datetime.timedelta(seconds=random_seconds)
            formatted_timestamp = timestamp.strftime("%Y-%m-%d %H:%M:%S")
            # Generate ID_Date in YYYYMMDD format
            id_date = timestamp.strftime("%Y%m%d")

            writer.writerow([user_id, feature, interaction_count, total_time, formatted_timestamp, id_date])

print("CSV file 'user_feature_interactions.csv' generated successfully.")
```

```python
import csv
import random
import datetime

# Parameters
num_users = 1000
# Probability that a user churns (and thus gets a subscription record)
churn_probability = 0.427

# Define the period for the first connection dates (e.g., first quarter of 2025)
first_connection_start = datetime.datetime(2025, 1, 1)
first_connection_end = datetime.datetime(2025, 3, 31)
days_range_fc = (first_connection_end - first_connection_start).days

# Open the CSV file for writing
with open("subscriptions.csv", "w", newline="", encoding="utf-8") as csvfile:
    writer = csv.writer(csvfile)

    # Write header row including the new ID_Date column (derived from start subscription date)
    writer.writerow([
        "Subscription_ID",
        "User_ID",
        "User_First_Connection_Date",
        "Start_Subscription_Date",
        "ID_Date",
        "End_Subscription_Date"
    ])

    subscription_id = 1
    # Iterate over users and only create subscription records for those with churn = 1
    for user_id in range(1, num_users + 1):
        if random.random() < churn_probability:
            # Generate a first connection date randomly within the period
            fc_offset = random.randint(0, days_range_fc)
            first_connection_date = first_connection_start + datetime.timedelta(days=fc_offset)
            first_connection_str = first_connection_date.strftime("%Y-%m-%d")

            # Generate a start subscription date between first connection date and +60 days
            start_offset = random.randint(0, 60)
            start_subscription_date = first_connection_date + datetime.timedelta(days=start_offset)
            start_subscription_str = start_subscription_date.strftime("%Y-%m-%d")

            # Generate ID_Date based on the start subscription date (in YYYYMMDD format)
            id_date = start_subscription_date.strftime("%Y%m%d")

            # Generate an end subscription date after the start date (e.g., 1 to 30 days after)
            if random.random() < 0.86:
                end_subscription_str = ""
            else:
                end_offset = random.randint(1, 30)
                end_subscription_date = start_subscription_date + datetime.timedelta(days=end_offset)
                end_subscription_str = end_subscription_date.strftime("%Y-%m-%d")

            writer.writerow([
                subscription_id,
                user_id,
                first_connection_str,
                start_subscription_str,
                id_date,
                end_subscription_str
            ])
            subscription_id += 1

print("CSV file 'subscriptions.csv' generated successfully.")
```

```python
import csv
import random
import datetime

# Number of users
num_users = 1000

# Define the ordered funnel stages and their exclusive probabilities (sum = 1)
stage_probs = [
    ("Page View",                 0.70),    # Stage 1
    ("CTA Click",                 0.15),    # Stage 2
    ("Signup",                    0.06),    # Stage 3
    ("First Data Connection",     0.027),   # Stage 4
    ("First Dashboard Created",   0.0126),  # Stage 5
    ("Features Engagement",       0.04032), # Stage 6
    ("Paid Conversion",           0.01008)  # Stage 7
]

# Build cumulative thresholds
cum = 0.0
cum_probs = []
for idx, (name, p) in enumerate(stage_probs, start=1):
    cum += p
    cum_probs.append((idx, name, cum))

# Timestamp window for initial page view
start_date = datetime.datetime(2025, 1, 1)
duration_days = 90
total_seconds = duration_days * 24 * 3600

# Seed for reproducibility
random.seed(42)

# Simulate average usage per feature for each user as integers
avg_usage = {
    user_id: (
        random.randint(0, 100),  # Avg Feature A usage (integer)
        random.randint(0, 100),  # Avg Feature B usage (integer)
        random.randint(0, 100)   # Avg Feature C usage (integer)
    )
    for user_id in range(1, num_users + 1)
}

# Write tidy funnel data with average usage columns (integers)
with open("funnel_tidy.csv", "w", newline="", encoding="utf-8") as f:
    writer = csv.writer(f)
    # Header including ID_Date and average usage columns
    writer.writerow([
        "User_ID", "Stage_Order", "Stage_Name", "Stage_Date", "ID_Date",
        "Avg_Feature_A_Usage", "Avg_Feature_B_Usage", "Avg_Feature_C_Usage"
    ])

    for user_id in range(1, num_users + 1):
        # 1) Pick the user's final stage
        r = random.random()
        for order, name, threshold in cum_probs:
            if r < threshold:
                final_order = order
                break

        # 2) Simulate the date they hit stage 1 (Page View)
        base_secs = random.randint(0, total_seconds)
        stage_date = start_date + datetime.timedelta(seconds=base_secs)

        # 3) For each stage up to final_order, emit a row and advance date
        for order, name, _ in cum_probs:
            if order > final_order:
                break
            # Format dates
            stage_date_str = stage_date.strftime("%Y-%m-%d")
            id_date = stage_date.strftime("%Y%m%d")

            # Retrieve integer average usages
            avgA, avgB, avgC = avg_usage[user_id]

            # Write the row
            writer.writerow([
                user_id,
                order,
                name,
                stage_date_str,
                id_date,
                avgA,
                avgB,
                avgC
            ])

            # Advance the date by 1â€“7 days for the next stage
            stage_date += datetime.timedelta(
                days=random.randint(1, 7),
                seconds=random.randint(0, 86400)
            )

print("CSV file 'funnel_tidy' generated successfully.")
```
